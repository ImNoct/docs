# Резисторы

```python
# резисторы
path = 'tables/sprav2006/резисторы/Характеристика надежности и справочные данные отдельных групп резисторов.csv'
df1 = pd.read_csv(path)
display(df1)
```

```python
def parse(path):
    result = []

    lines = open(path).readlines()
    i = 0

    last_group = ''
    last_subgroup = ''

    while i < len(lines):
        line = lines[i].strip()

        if not line:
            i += 1
            continue

        if line[:2] == '##':
            last_subgroup = line[2:].strip()
            i += 1

            continue

        if line[:1] == '#':
            last_group = line[1:].strip()
            last_subgroup = ''
            i += 1
            continue

        f1 = lines[i].strip()
        i += 1
        f2 = lines[i].strip()
        i += 1        
        f3 = lines[i].strip()
        i += 1        
        f4 = lines[i].strip()
        i += 1        
        f5 = lines[i].strip()
        i += 1        
        f6 = lines[i].strip()
        i += 1

        result.append(
            {
                "Группа изделия" : last_group,
                "Подгруппа изделия" : last_subgroup,
                "Тип изделия" : f1,
                "d, шт." : f2,
                "λб⋅10^6, 1/ч" : f3,
                "Тн.м, тыс. ч" : f4,
                "Тр.γ, тыс. ч (γ = 95%)" : f5,
                "Тхр, лет" : f6
            }
        )
    
    return result

r = parse('/home/admen/Desktop/pdf/source_preparation/tables/sprav2006/резисторы/типы_таблица.txt')
df = pd.DataFrame(r)
df = df.replace(['-', '–', ''], float('nan'))
display(df)
df.to_csv('tables/sprav2006/резисторы/Характеристика надежности и справочные данные отдельных типов резисторов.csv',index=False)
```
